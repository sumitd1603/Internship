Data Collection Phase: In this section, you need to scrape images from e-commerce portal, Amazon.com. The clothing categories used for scraping will be:
-	Sarees (women)
-	Trousers (men)
-	Jeans (men)


```python
# Importing libraries
import selenium
import time
from bs4 import BeautifulSoup
import os
import requests
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.chromium import service
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
```

```python
# method to store all item's ursl

links=[]        # empty list to store urls
def urls(item, path):
    
    driver = webdriver.Chrome(r'D:\Machine Learning Course\Internship\chromedriver_win32 (1)\chromedriver.exe')
    driver.get('https://www.amazon.in/')
    
    driver.find_element(By.ID,"twotabsearchtextbox").clear()
    time.sleep(2)
    driver.find_element(By.ID,"twotabsearchtextbox").send_keys(item,Keys.ENTER)
    time.sleep(2)
           
    start_page = 0
    end_page = 7
    for i in range(start_page,end_page+1):
        sar=driver.find_elements(By.XPATH,path)
        for j in sar:
            image=j.get_attribute('src')
            links.append(image)
        
        nxt_button=driver.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        driver.get(nxt_button.get_attribute('href'))
        time.sleep(5)
```

```python
# Storing all links of Sarees into a list
urls("Sarees", "//div[@class='a-section aok-relative s-image-tall-aspect']/img")
```

```python
# Storing links into variable
sarees=links
print(len(sarees))
```

```python
# Storing all links of Jeans into a list
links=[]
urls("Jeans men", "//div[@class='a-section aok-relative s-image-tall-aspect']/img")
```

```python
# Storing links into variable
Jeans=links
print(len(Jeans))
```

```python
# Storing all links of Trousers into a list
links=[]
urls("Trousers men", "//div[@class='a-section aok-relative s-image-tall-aspect']/img")
```

```python
# Storing links into variable
Trousers=links
print(len(Trousers))
```

```python
# using the list of links ,save the image into our local path

def saving_images(folder,list1,item):
    current = os.getcwd()
    new_path = os.path.join(current,folder)
    if not os.path.exists(new_path):
        os.makedirs(new_path)
        
    for i,link in enumerate(list1):
        try:
            response = requests.get(link)
            print("downloading {} of {}".format(i+1,len(list1)))
            with open(folder+"\\"+item+"_{}.jpg".format(i+1),"wb") as file:
                file.write(response.content)
        except:
            pass
            print("Error occured. Continuing...")
```

```python
saving_images("Saree women",sarees[:400],"saree")
```

```python
saving_images("Jeans men",Jeans[:400],"jean")
```

```python
saving_images("Trousers men",Trousers[:400],"trouser")
```
